**Possuir ou não veículo é estatisticamente mais signifcante do que tipo de renda e nível de escolaridade.**

````python
sns.pointplot(x = 'posse_de_veiculo', y = 'renda',data = df, dodge = True, errorbar=('ci', 90))
````

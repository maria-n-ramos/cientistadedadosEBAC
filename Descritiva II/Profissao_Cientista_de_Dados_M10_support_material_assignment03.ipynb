````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('https://raw.githubusercontent.com/maria-n-ramos/cientistadedadosEBAC/main/Descritiva%20II/Profissao-%20Cientista%20de%20Dados_M10_support%20material.csv')
df.head()
````

````python
df['data_T'] = pd.to_datetime(df['data_ref'])
````

````python
_, ax = plt.subplots(figsize=(20, 10))

ax = sns.countplot(x = 'data_T',hue = 'mau', data = df)

tick_labs = df['data_T'].map(lambda ts: ts.strftime("%m-%Y")).unique()

ticks = ax.set_xticks(list(range(df['data_T'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation = 0)

plt.legend(bbox_to_anchor = (1.05, 1), loc = 2, borderaxespad = 0)
````

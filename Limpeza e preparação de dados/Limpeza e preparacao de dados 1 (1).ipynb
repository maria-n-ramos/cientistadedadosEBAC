````python
import pandas as pd
import requests
import numpy as np
# 1) seu código aqui
sinasc = pd.read_csv('SINASC_RO_2019.csv')
print(sinasc.shape)
sinasc.drop_duplicates().shape
# Não há duplicados
````

````python
# 2) seu código aqui
sinasc.isna().sum()
````

````python
# 3) seu código aqui
#Ok, no item anterior você deve ter achado pouco prático ler a informação de tantas variáveis, muitas delas nem devem ser interesantes. Então crie uma seleção dessa base somente com as colunas que interessam. São elas:
#
#['LOCNASC', 'IDADEMAE', 'ESTCIVMAE', 'ESCMAE', 'QTDFILVIVO', 
# 'GESTACAO', 'GRAVIDEZ', 'CONSULTAS', 'APGAR5'] 

base = pd.DataFrame(
    columns=['LOCNASC', 'IDADEMAE', 'ESTCIVMAE', 'ESCMAE', 'QTDFILVIVO', 
 'GESTACAO', 'GRAVIDEZ', 'CONSULTAS', 'APGAR5'],
)
base ['LOCNASC'] = sinasc ['LOCNASC']
base ['IDADEMAE'] = sinasc ['IDADEMAE']
base ['ESTCIVMAE'] = sinasc['ESTCIVMAE']
base ['ESCMAE'] = sinasc['ESCMAE']
base ['QTDFILVIVO'] = sinasc['QTDFILVIVO']
base ['GESTACAO'] = sinasc['GESTACAO']
base ['GRAVIDEZ'] = sinasc['GRAVIDEZ']
base ['CONSULTAS'] = sinasc['CONSULTAS']
base ['APGAR5'] = sinasc['APGAR5']
base
````

````python
base.isna().sum()
````

````python
# 4) seu código aqui
base2 = base.dropna(subset=['APGAR5']) #nova base com remoção dos missings em APGAR5
base2.isna().sum()
````

````python
# 5) seu código aqui
#observe que as variáveis ['ESTCIVMAE', 'CONSULTAS'] possuem o código 9, que significa ignorado. 
#Vamos assumir que o não preenchido é o mesmo que o código 9.
base2['APGAR5'].fillna(9, inplace = True)
#base2 #não tá trocando o 9 por NaN
base2['APGAR5'].value_counts()
````

````python
# 6) Seu código aqui
#Substitua os valores faltantes da quantitativa (QTDFILVIVO) por zero.
base2['QTDFILVIVO'].fillna(0,inplace = True)
base2.isna().sum()
````

````python
# 7) seu código aqui
#Das restantes, decida que valore te parece mais adequado 
#(um 'não preenchido' ou um valor 'mais provável' como no item anterior) 
#e preencha. Justifique.
base2['ESTCIVMAE'].fillna('não preenchido', inplace = True)#Estado civil
base2['ESCMAE'].fillna('não preenchido', inplace = True)#Escolaridade

medianaQTDFILVIVO = base2['QTDFILVIVO'].median()
base2['QTDFILVIVO'].fillna(medianaQTDFILVIVO, inplace = True)

base2['GESTACAO'].fillna('não preenchido', inplace = True)#Semanas de gestação

base2['GRAVIDEZ'].fillna('não preenchido', inplace = True)#Tipo de gravidez
base2.head()
````

````python
# 8) seu código aqui
#converter pra inteiro pra rodar a sequencia abaixo
#base2['APGAR5'] = base2['APGAR5'].astype('float64')
#base2 = base2['APGAR5'].apply(lambda x: float(x))
````

````python
# 8) seu código aqui
#O Apgar possui uma classificação indicando se o bebê passou por asfixia:

#Entre 8 e 10 está em uma faixa 'normal'.
#Entre 6 e 7, significa que o recém-nascido passou por 'asfixia leve'.
#Entre 4 e 5 significa 'asfixia moderada'.
#Entre 0 e 3 significa 'asfixia severa'.
#Crie uma categorização dessa variável com essa codificação e calcule as frequências dessa categorização

base2.loc[(base2['APGAR5'] >= 8) & (base2['APGAR5'] <= 10) ,'classificacao_apgar5'] = 'normal'
base2.loc[(base2['APGAR5'] >= 6) & (base2['APGAR5'] <= 7),'classificacao_apgar5'] = 'asfixia leve'
base2.loc[(base2['APGAR5'] >= 4) & (base2['APGAR5'] <= 5),'classificacao_apgar5'] = 'asfixia moderada'
base2.loc[base2['APGAR5'] <=3,'classificacao_apgar5'] = 'asfixia severa'
base2.head()
````

````python
# 9) seu código aqui
#Renomeie as variáveis para que fiquem no snake case, ou seja, 
#em letras minúsculas, com um underscore entre as palávras. 
#Dica: repare que se você não quiser criar um dataframe novo, 
#você vai precisar usar a opção inplace = True
['LOCNASC', 'IDADEMAE', 'ESTCIVMAE', 'ESCMAE', 'QTDFILVIVO', 
 'GESTACAO', 'GRAVIDEZ', 'CONSULTAS', 'APGAR5']
mapeamento = {'LOCNASC': 'local_de_nascimento'}
base2.rename(columns = mapeamento, inplace = True)

mapeamento = {'IDADEMAE': 'local_de_nascimento'}
base2.rename(columns = mapeamento, inplace = True)

mapeamento = {'ESTCIVMAE': 'estado_civil'}
base2.rename(columns = mapeamento, inplace = True)

mapeamento = {'ESCMAE': 'escolaridade'}
base2.rename(columns = mapeamento, inplace = True)

mapeamento = {'QTDFILVIVO': 'quantidade_de_filhos_vivos'}
base2.rename(columns = mapeamento, inplace = True)

mapeamento = {'GESTACAO': 'gestacao'}
base2.rename(columns = mapeamento, inplace = True)

mapeamento = {'GRAVIDEZ': 'gravidez'}
base2.rename(columns = mapeamento, inplace = True)

mapeamento = {'CONSULTAS': 'consultas'}
base2.rename(columns = mapeamento, inplace = True)

mapeamento = {'APGAR5': 'apgar5'}
base2.rename(columns = mapeamento, inplace = True)
base.head()
````

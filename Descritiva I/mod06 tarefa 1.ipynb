```python
import pandas as pd
import requests
import numpy as np
import seaborn as sns
# 1) seu código aqui
sinasc = pd.read_csv('SINASC_RO_2019.csv')
print(sinasc.shape)
sinasc.drop_duplicates().shape
# Não há duplicados
```

```python
import matplotlib.pyplot as plt

# %matplotlib inline
%matplotlib notebook
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
plt.close('all')

plt.rc('figure', figsize = (8, 6))

fig, axes = plt.subplots(3, 2)

sns.countplot(ax = axes[0, 0], x = 'APGAR1', data = sinasc)
sns.countplot(ax = axes[0, 1], x = 'APGAR5', data = sinasc)
sns.countplot(ax = axes[1, 0], x = 'SEXO', data = sinasc)
sns.countplot(ax = axes[1, 1], x = 'ESTCIVMAE', data = sinasc)
sns.countplot(ax = axes[2, 0], x = 'GESTACAO', data = sinasc).tick_params(axis = 'x', rotation = 30)
plt.subplots_adjust(wspace = 0.2, hspace = 0.2)
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.close('all')
sns.barplot(x = 'CONSULTAS', y = 'APGAR5', data = sinasc)
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.close('all')

lista = [['SEXO', 'GESTACAO'], ['ESTCIVMAE', 'CONSULTAS']]

plt.close('all')
plt.rc('figure', figsize = (10, 10))
fig, axes = plt.subplots(2, 2)

for i in range(2):
    for j in range(2):
        sns.countplot(ax = axes[i, j], x = lista[i][j], data = sinasc)
        if i == 0 and j == 1:
            axes[i, j].tick_params(axis = 'x', rotation = 90)
            plt.xlabel("Gestação")
#plt.xlabel("teste")
plt.subplots_adjust(wspace = 0.25, hspace = 1)
```
